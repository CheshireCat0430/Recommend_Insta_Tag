```python
import tweepy
from tweepy import OAuthHandler
import json
import wget
import os
import cv2

auth = tweepy.OAuthHandler(CONSUMER_KEY,CONSUMER_SECRET)
auth.set_access_token(ACCESS_TOKEN,ACCESS_TOKEN_SECRET)
api = tweepy.API(auth,wait_on_rate_limit=True)
```

```python
'''
file_list = os.listdir('./querylist')
file_name = []
for file in file_list:
    if file.count(".") == 1: 
        name = file.split('.')[0]
        file_name.append(name)
    else:
        for k in range(len(file)-1,0,-1):
            if file[k]=='.':
                file_name.append(file[:k])
                break
'''
file_name = open('./querylist/labels.txt','r').read().split('\n')

for f in file_name:
  save_keyword = f
  keyword = f #저장용 이름은 키워드(분류)로 하되, 실제 검색에서는 다양한 단어를 찾아봐주길 바래 회같은 경우는 회로 검색하는거보다 생선이름으로 검색하는게 정확하고 많이나옴
  txtfile = open('./querylist/'+f+'.txt','r')
  query = txtfile.read().split('\n')
  result = []
  result_temp = []
  pic = []
  txt = []
  max_tweets = 500
  last_id = -1
  
#search
  for q in query:
    while len(result_temp) < max_tweets:
      count = max_tweets - len(result_temp)
      try:
        new_tweets = api.search(q=q+" filter:images",tfilter=" -filter:retweets",resultType="mixed",count=count,max_id=str(last_id-1))
        if not new_tweets:
          break
        result_temp.extend(new_tweets)
        last_id = new_tweets[-1].id
      except tweepy.TweepError as e:
        break
    result.extend(result_temp)
    print(len(result_temp))
    result_temp=[]
    last_id = -1

#save
  addr = './Image/image/'+save_keyword+'/'
  addr2 = './Image/image_re/'+save_keyword+'/'

  if not os.path.isdir(addr):
    os.mkdir(addr)
  if not os.path.isdir(addr2):
    os.mkdir(addr2)

  i = 1

  for a in result:
    try:
      chk = [q for q in query if q in a.text]
      if keyword in a.text or len(chk)>0:
        
        for count in range(0,len(a.extended_entities['media'])):
          if a.extended_entities['media'][count]['type'] == "photo" and a.extended_entities['media'][count]['media_url'] not in pic:
            
            pic.append(a.extended_entities['media'][count]['media_url'])
            wget.download(pic[i],addr+save_keyword+str(i)+"_"+str(count+1)+".jpg")
            image = cv2.imread(addr+save_keyword+str(i)+"_"+str(count+1)+".jpg",cv2.IMREAD_COLOR)
            image_resized = cv2.resize(image, (150,150))
            cv2.imwrite(addr2+save_keyword+str(i)+"_"+str(count+1)+".jpg",image_resized)
            i = i+1


    except:
      print(end='')
```
